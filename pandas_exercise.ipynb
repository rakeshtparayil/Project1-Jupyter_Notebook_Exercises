# exercise 17 
most & least frequent MPAA Rating in the dataset in terms of occurrences

```python
# most frequent value
freq = movies['mpaa_rating'].value_counts()
freq

# most_freq_value_MPAA = freq.index[3]
most_freq_mpaa = freq.idxmax()
most_freq_mpaa

# # least frequent value
least_freq_mpaa = freq.idxmin()
least_freq_mpaa

print(freq)
print(most_freq_mpaa)
print(least_freq_mpaa)
```

# exercise 18 
most & least expensive Genre (take an average of all Budget measures grouped by Genre - use the groupBy() method)

```python
#average of all budget group by
average_budget = movies.groupby('genre')['budget'].mean()
average_budget

most_expensive_genre = average_budget.idxmax()
most_expensive_genre

least_expensive_genre = average_budget.idxmin()
least_expensive_genre

print("the most expensive genre is",most_expensive_genre)
print("the least expensive genre is",least_expensive_genre)
```
