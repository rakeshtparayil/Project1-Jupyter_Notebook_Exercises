# Exercise 6 
use the fillna() method to replace the missing values of Rating with their median

```python
movies.fillna({ 'rating': movies['rating'].median() }, inplace=True)
print(movies.isnull().sum())
```

# Exercise 7 
use the fillna() method to replace the missing values of rating_count with their median

```python
movies.fillna({ 'rating_count': movies['rating_count'].median() }, inplace=True)
print(movies.isnull().sum())
```
